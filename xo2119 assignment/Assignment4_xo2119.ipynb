From the data, I can plot line graphs. But the Age and Client ID don't have any relationship. The same reason for the scatter points graph. Both of them try to connect the Age and Client ID which is meaningless.

Bar charts are useful when we would like to visualize categorical distributions. But in the table, the category is only age ranges from 1 to 95. It will be too many bars in the plot. So it is not a good choice.

Histograms is the best choice to visualize the age distribution of itsclientsâ€™ database.It can show the density distribution of different ages. And it doesn't related to the Client ID. we can also group the age into four parts of 0-30, 30-50, 50-70, 70-100. And analyze them separately.

```python
age.select('Age').hist(bins = np.arange(-5,105,5),width = 10)
```

Insights: 
People of age from 10-15 and 35-40 are most likely to buy the insurance. Maybe in age 10-15, people are easy to get injured in their adolescence. And in age 35-40, people will pay more attention to their healthy situation. 

People of age from 65-100 have the lowest potential to buy the insurance. So they should not be the main costomers. 

People of age from 10-25 and 30-50 have the biggest potential of purchasing the insurance. So they should be the main target customers.

```python
age.select('Age').hist(bins = np.arange(0,32,2),width = 5)
plt.title('Age from 0 to 30');
```

The distribution from 0-30 is not uniform. 28-30 has the highest density.

```python
age.select('Age').hist(bins = np.arange(30,52,2),width = 5)
plt.title('Age from 30 to 50');
```

The distribution from 38-46 is uniform. 34-36 has the highest density.

```python
age.select('Age').hist(bins = np.arange(50,72,2),width = 5)
plt.title('Age from 50 to 70');
```

The distribution from 52-56 and 60-66 are uniform. 

```python
age.select('Age').hist(bins = np.arange(70,102,2),width = 5)
plt.title('Age from 70 to 100');
```

74-76 and 78-80 have the highest density.
